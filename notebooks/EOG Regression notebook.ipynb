**EOG activity removal using linear regression**

Data is loaded from a mat file containing EEG and EOG data.
EOG activity is subtracted from the EEG activity.

```python
from scipy.io import loadmat
import os
import sys
pyautomagic_dir = os.path.abspath(os.path.dirname(os.getcwd()))
sys.path.append(pyautomagic_dir)
from pyautomagic.preprocessing.perform_EOG_regression import perform_EOG_regression

sfreq = 250
a=loadmat('../tests/test_data/A01ChannelData.mat')
eeg=a['ChannelData']
n_chans = 22
EEG = eeg[0:20,:]
EOG = eeg[20:23]
clean_eeg=perform_EOG_regression(EEG, EOG)
```

```python
import matplotlib.pyplot as plt
plt.ylim(-100,100)
plt.ylabel('Voltage (microvolts)')
plt.plot(EEG[1,:])
plt.show()

plt.ylim(-100,100)
plt.ylabel('Voltage (microvolts)')
plt.plot(clean_eeg[1,:])
plt.show()
```
